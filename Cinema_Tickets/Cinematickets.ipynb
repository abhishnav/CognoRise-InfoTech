```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/cinemaTicket_Ref.csv')
df
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
# Droping columns month and day

df.drop(['month','day'],inplace=True,axis=1)
df
```

```python
df['date']=df['date'].str.replace('-','')
df['date']=df['date'].astype(int)
```

```python
df.corr()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
df.drop(['ticket_use'],inplace=True,axis=1)
df
```

```python
# Splitting data into x and y

x=df.drop(['total_sales'],axis=1)
x
```

```python
y=df['total_sales']
y
```

```python
# Regression plot of highly correlated features with target

sns.regplot(x=df['tickets_sold'],y=y,color='blue')
plt.title('Tickets_sold - Total_sales',color='black')
```

```python
sns.regplot(x=df['show_time'],y=y,color='blue')
plt.title('Show_time - Total_sales',color='black')
```

```python
# Regression plot of least correlated features with target

sns.regplot(x=df['cinema_code'],y=y,color='red')
plt.title('Cinema_code - Total_sales',color='black')
```

```python
# Mutual Information Gain

from sklearn.feature_selection import mutual_info_classif
mutual_info=mutual_info_classif(x,y,random_state=42)
mutual_info
```

```python
mutual_info=pd.Series(mutual_info)
mutual_info.index=x.columns
mutual_info.sort_values(ascending=False)
```

```python
mutual_info.sort_values(ascending=False).plot.bar()
```

```python
# Regression plot

sns.regplot(x=df['tickets_out'],y=y,color='green')
plt.title('Tickets_out - Ticket_sales',color='black')
```

```python
# tickets_out can be dropped

df.drop(['tickets_out'],inplace=True,axis=1)
```

```python
# Splitting imto training and testing data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# Model-Linear Regression

from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
# Model Performance Evaluation

from sklearn.metrics import mean_absolute_percentage_error,r2_score,mean_absolute_error,mean_squared_error
print('MAPE is',mean_absolute_percentage_error(y_test,y_pred))
print('R2 SCORE is',r2_score(y_test,y_pred))
print('MAE is',mean_absolute_error(y_test,y_pred))
print('MSE is',mean_squared_error(y_test,y_pred))
```

```python
# Model-Decison Tree

from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(criterion='absolute_error')
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
y_pred
```

```python
# Model Evaluation

from sklearn.metrics import mean_absolute_percentage_error,r2_score,mean_absolute_error,mean_squared_error
print('MAPE is',mean_absolute_percentage_error(y_test,y_pred))
print('R2 SCORE is',r2_score(y_test,y_pred))
print('MAE is',mean_absolute_error(y_test,y_pred))
print('MSE is',mean_squared_error(y_test,y_pred))
```

Decision Tree shows more performance.
